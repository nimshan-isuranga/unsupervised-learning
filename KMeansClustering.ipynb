```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Wine.csv")
data.head()
```

```python
data.iloc[:,0].values
```

```python
x=data.iloc[:,1:]
```

```python
kmc=KMeans(n_clusters=3)
```

```python
kmc.fit(x)
```

```python
kmc.labels_
```

```python
# those are representing clusters not the actual class
```

```python
kmc.inertia_
```

```python
D=[]
K=range(1,15)
for k in K:
  model=KMeans(n_clusters=k)
  model.fit(x)
  D.append(model.inertia_)
```

```python
plt.plot(K,D,"bx-")
plt.show()
```
