```python
bkt=data_ger.groupby(["InvoiceNo","Description"])["Quantity"].sum().unstack().fillna(0)
bkt
```

```python
bkt=bkt.astype(int)
```

```python
# function to quantity available 
def encode_units(x):
  if x>=1:
    return 1 
  else:
    return 0
```

```python
bkt=bkt.applymap(encode_units)
```

```python
ap_results=apriori(bkt,min_support=0.05,use_colnames=True)
ap_results
```

```python
rules=association_rules(ap_results,metric="lift",min_threshold=1)
rules
```

```python
rules[(rules["lift"]>=3) & (rules["confidence"]>=0.5) ]
```
